# VDI Score Analysis
Analysis of VDI scores issues by BART Large MNLI using max entailment probability of top 5 chunks for RIL (2014-2018)

```python
gold_standard = [50, 51, 52, 52, 53]
predicted = [66, 68, 64, 66, 63]
```

```python
plt.title('VDI Score Analysis for RIL (2014-2018)')
plt.scatter(gold_standard, predicted)
plt.xlabel('Gold Standard')
plt.ylabel('Predicted')
plt.show()
```
